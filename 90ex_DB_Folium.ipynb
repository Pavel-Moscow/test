```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/api/dataset_part_2.csv')
df.head(5)
```

```python
spacex_df = df [['LaunchSite','Longitude','Latitude', 'Class']]
site_df = spacex_df.groupby(['LaunchSite'], as_index=False).first()
site_df.drop('Class', axis = 1, inplace = True)
site_df
```

I created map with NASA coordinates to have centered map with NASA head office point

```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

Folium map with Launch site coordinates:

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=4)
# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label
site_points = folium.map.FeatureGroup()
for lat, long, site in zip(site_df['Latitude'], site_df['Longitude'], site_df['LaunchSite']):
    site_points.add_child(
        folium.features.Circle([lat, long],
                                radius = 60,
                                color = 'darkblue',
                                fill = True,
                                fill_color = 'lightgreen',
                                fill_opacity = 0.5,
                                popup = site)
    )

site_names = folium.map.FeatureGroup()    
for lat, long, site in zip(site_df['Latitude'], site_df['Longitude'], site_df['LaunchSite']):
    site_names.add_child(
    folium.features.Marker([lat, long],
                         icon = DivIcon(icon_size=(20,20),icon_anchor=(0,0), 
                                        html='<div style="font-size: 12; color:#dc143c;"><b>%s</b></div>' % site)
                         )
    )
    
    
site_map.add_child(site_points)
site_map.add_child(site_names)
```

Creating of marker cluster for our map:

```python
# Add marker_cluster to current site_map
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, 
# e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
for index, record in spacex_df.iterrows():
    # TODO: Create and add a Marker cluster to the site map
    marker = folium.Marker([spacex_df.loc[index, 'Latitude'], spacex_df.loc[index, 'Longitude']],
                          popup = ('Outcome = ' + str(spacex_df.loc[index, 'Class'])),
                          icon=folium.Icon(color = spacex_df.loc[index, 'result'])
                          )
                          
    marker_cluster.add_child(marker)
                          
site_map
```

```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
city_CCAFS_SLC_40 = calculate_distance(28.561857, -80.577366, 28.4, -80.603)
city_KSC_LC_39A = calculate_distance(28.608058, -80.603956, 28.61, -80.81)
city_VAFB_SLC = calculate_distance(34.632093, -120.610829, 34.6501, -120.458)
print('city - CCAFS_SLC_40:', city_CCAFS_SLC_40, 'city - KSC_LC_39A:', city_KSC_LC_39A, 'city - VAFB_SLC:', city_VAFB_SLC)
```

```python
city_dist = {'city_CCAFS_SLC_40':([28.4, -80.603], 18.2),
                    'city_KSC_LC_39A':([28.61, -80.81], 20.1), 'city_VAFB_SLC':([34.6501, -120.458], 14.1)}
```

```python
city_dist_circle =  folium.map.FeatureGroup()
for key in city_dist.keys():
    city_dist_circle.add_child(
        folium.features.Circle(city_dist[key][0],
                                radius = 1000,
                                color = 'blue',
                                fill = True,
                                fill_color = 'beige',
                                fill_opacity = 0.8,
                                )
    )


city_distance_marker = folium.map.FeatureGroup()  
for key in city_dist.keys():
    city_distance_marker.add_child(
        folium.features.Marker(
            city_dist[key][0],
            icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0), 
            html='<div style="font-size: 12; color:#4B0082;"><b>%s</b></div>' % (str(city_dist[key][1]) + ' km',)
                        )
        )
    )
    
site_map.add_child(city_dist_circle)
site_map.add_child(city_distance_marker)

city_line1 = folium.PolyLine(locations=[[28.561857, -80.577366], [28.4, -80.603]], weight=3, color = 'green')
site_map.add_child(city_line1)

city_line2 = folium.PolyLine(locations=[[28.608058, -80.603956], [28.61, -80.81]], weight=3, color = 'green')
site_map.add_child(city_line2)

city_line3 = folium.PolyLine(locations=[[34.632093, -120.610829], [34.6501, -120.458]], weight=3, color = 'green')
site_map.add_child(city_line3)

site_map
```
